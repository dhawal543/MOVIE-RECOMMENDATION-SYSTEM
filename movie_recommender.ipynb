## Movie Recommender System

```python
import numpy as np 
import pandas as pd

```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv') 
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies.shape
```

```python
credits.shape
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.shape
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
import ast
```

```python
def convert(text):
    L=[]
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
movies.shape
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ", ""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_movies = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new_movies.head()
```

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x: " ".join(x))
new_movies.head()
```

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x : x.replace(",", ""))
new_movies['tags'] = new_movies['tags'].apply(lambda x : x.lower())
```

```python
new_movies.head(30)
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
def filter_stop_words(text):
    filtered_sentence = []
    for w in text.split():
        if w not in stop_words:
            filtered_sentence.append(w)
    return filtered_sentence
```

```python
new_movies['tags'] = new_movies['tags'].apply(filter_stop_words)
```

```python
new_movies.head(10)
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
def root_word(text):
    L = []
    for w in text:
        L.append(ps.stem(w))
    return L
```

```python
new_movies['tags'] = new_movies['tags'].apply(root_word)
```

```python
new_movies.tags.head(30)
```

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x: " ".join(x))
new_movies.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(max_features=10000)
```

```python
vectors = vectorizer.fit_transform(new_movies['tags'])
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
print(similarity)
```

```python
import difflib
```

```python
def recommend(movie):
    index = new_movies[new_movies['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:11]:
        print(new_movies.iloc[i[0]].title)
        
```

```python
movie_name = input('Enter the movie name: ')
```

```python
list_of_all_titles = new_movies['title'].tolist()
```

```python
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
```

```python
close_match = find_close_match[0]
```

```python
recommend(close_match)
```
